```python
# ðŸ“Š ESTADÃSTICAS RÃPIDAS DEL DATASET
import pandas as pd

# Cargar dataset
df = pd.read_csv('listings_nuevo.csv')

print("ðŸ“Š ESTADÃSTICAS GENERALES")
print("=" * 40)
print(f"ðŸ“‹ Total registros: {len(df):,}")
print(f"ðŸ“Š Total columnas: {len(df.columns)}")
print(f"ðŸŒ Ciudades: {', '.join(df['city'].unique())}")

print(f"\nðŸ™ï¸ DISTRIBUCIÃ“N POR CIUDAD:")
city_counts = df['city'].value_counts()
for city, count in city_counts.items():
    percentage = (count / len(df)) * 100
    print(f"  â€¢ {city:9}: {count:5,} registros ({percentage:4.1f}%)")

print(f"\nðŸ  TIPOS DE HABITACIÃ“N:")
room_types = df['room_type'].value_counts()
for room_type, count in room_types.items():
    percentage = (count / len(df)) * 100
    print(f"  â€¢ {room_type:15}: {count:5,} registros ({percentage:4.1f}%)")

print(f"\nðŸ’° ESTADÃSTICAS DE PRECIOS:")
price_stats = df['price'].describe()
print(f"  â€¢ Precio promedio: ${price_stats['mean']:.2f}")
print(f"  â€¢ Precio mediano: ${price_stats['50%']:.2f}")
print(f"  â€¢ Rango de precios: ${price_stats['min']:.2f} - ${price_stats['max']:.2f}")

print(f"\nðŸ“… ESTANCIAS MÃNIMAS:")
nights_stats = df['minimum_nights'].describe()
print(f"  â€¢ Promedio: {nights_stats['mean']:.1f} noches")
print(f"  â€¢ Mediana: {nights_stats['50%']:.0f} noches")
print(f"  â€¢ MÃ¡ximo: {nights_stats['max']:.0f} noches")

print(f"\nðŸ” CALIDAD DE DATOS:")
print(f"  â€¢ Valores nulos totales: {df.isnull().sum().sum():,}")
print(f"  â€¢ Columnas con nulos: {(df.isnull().sum() > 0).sum()}")
print(f"  â€¢ Completitud promedio: {((df.notna().sum().sum()) / (len(df) * len(df.columns)) * 100):.1f}%")
```
